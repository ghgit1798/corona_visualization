# 결론
## 1. 해외 유입 확진자 수 증가에 따른 추가 조치 필요
- 국내 확진자 수가 안정기에 접어들면서, 외국에서 유입되는 확진자 수가 급격히 증가하는 것으로 파악되었습니다.
- 따라서 이들의 주요 방문지인 병원과 공항을 중심으로 철저한 방역이 필요할 것으로 보입니다.

## 2. 일별 확진자 수 감소에 따른 대중의 관심도 하락과 기온 상승
- 일별 확진자 수가 감소함에 따라 코로나 검색빈도도 줄어드는 것을 확인하였습니다.
- 서울을 기준으로 기온이 올라감에 따라 활동하기 좋은 날씨가 이어지고 있습니다.
- 따라서 코로나에 대한 사람들의 관심도가 줄어들고, 날씨가 좋아지면서 사회적 거리 두기가 느슨해질 가능성이 있다고 판단했습니다.

### 일별 확진자 수를 각 도시의 데이터프레임에 추가했습니다.

```python
data1 = [0]
data2 = [0]
data3 = [0]
data4 = [0]
for i in range(1,len(Daegu)):
    tmp = Daegu.iloc[i]['confirmed'] - Daegu.iloc[i-1]['confirmed']
    data1.append(tmp)
    
for i in range(1,len(Buk)):
    tmp = Buk.iloc[i]['confirmed'] - Buk.iloc[i-1]['confirmed']
    data2.append(tmp)
    
for i in range(1,len(Seoul)):
    tmp = Seoul.iloc[i]['confirmed'] - Seoul.iloc[i-1]['confirmed']
    data3.append(tmp)
    
for i in range(1,len(GG)):
    tmp = GG.iloc[i]['confirmed'] - GG.iloc[i-1]['confirmed']
    data4.append(tmp)

Daegu['date_confirmed'] = data1
Buk['date_confirmed'] = data2
Seoul['date_confirmed'] = data3
GG['date_confirmed'] = data4
Daegu.head()
```

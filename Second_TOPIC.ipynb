```python
data = city.sort_values(by=['date', 'confirmed'], ascending=False).head(8)
data = data[['province','confirmed','released']].reset_index()
data
```

```python
pie = Pie("각 도시별 확진자 대비 완치자 비율")
style = Style()
pie_style = style.add(
    label_pos="center",
    is_label_show=True,
    label_text_color=None,
    is_legend_show=False,
    is_random=True
)
x = 10
y = 30
for i in range(0,8):
    pie.add(
      "", [data.loc[i]["province"], str(data.index[i])+"_완치자"],
        [data.loc[i]['confirmed'],data.loc[i]['released']],
        center = [x,y], radius = [25,40], **pie_style
    )
    if i < 4:
        x = x+25
    if i == 3:
        x = 10
        y = 75
    if i > 3:
        x = x+25
pie.render()
pie
```

### 일별 확진자 수를 각 도시의 데이터프레임에 추가했습니다.

```python
data1 = [0]
data2 = [0]
data3 = [0]
data4 = [0]
for i in range(1,len(Daegu)):
    tmp = Daegu.iloc[i]['confirmed'] - Daegu.iloc[i-1]['confirmed']
    data1.append(tmp)
    
for i in range(1,len(Buk)):
    tmp = Buk.iloc[i]['confirmed'] - Buk.iloc[i-1]['confirmed']
    data2.append(tmp)
    
for i in range(1,len(Seoul)):
    tmp = Seoul.iloc[i]['confirmed'] - Seoul.iloc[i-1]['confirmed']
    data3.append(tmp)
    
for i in range(1,len(GG)):
    tmp = GG.iloc[i]['confirmed'] - GG.iloc[i-1]['confirmed']
    data4.append(tmp)

Daegu['date_confirmed'] = data1
Buk['date_confirmed'] = data2
Seoul['date_confirmed'] = data3
GG['date_confirmed'] = data4
Daegu.head()
```

## 이번엔 각 연령대별 성비가 어떻게 되는지 시각화 해보았습니다.

### 성별별로 환자 데이터 추출

```python
male = patient.loc[patient['sex']=='male']
female = patient.loc[patient['sex']=='female']
```

### 성별별로 뽑은 데이터의 연령대별 확진자수 파악

```python
# 0세의 성별을 파악하는 것은 의미가 없다고 판단. 삭제하였습니다.
m_age = male.age.value_counts()
m_age = m_age.drop('0s')
m_age
```

```python
# 0세의 성별을 파악하는 것은 의미가 없다고 판단. 삭제하였습니다.
f_age = female.age.value_counts()
f_age = f_age.drop('0s')
f_age
```

### 성별별 확진자 수를 다시 연령대를 기준으로 병합

```python
total_age = []
for i in range(len(m_age)):
    for j in range(len(f_age)):
        if m_age.index[i] == f_age.index[j]:
            total_age.append([m_age[i],f_age[j]])
age_index = m_age.index
total_age
```

### 데이터프레임을 만들면서 index도 다시 추가해주었습니다.

```python
age_index
```

```python
total_age = dict(zip(age_index, total_age))
total_age = pd.DataFrame(total_age)
```

```python
total_age = total_age[['10s','20s','30s','40s','50s','60s','70s','80s','90s']]
```

### 비교를 위한 전체 확진자 성비입니다.

```python
name = ['남성','여성']
data = gender.loc[gender['date']=='2020-04-13', 'confirmed']
pie = Pie("전체 확진자 성비")
pie.add("전체 확진자 성비", name, data, is_label_show=True)
pie
```

### 각 연령별 성비입니다.

```python
pie = Pie("각 연령별 성비")
style = Style()
pie_style = style.add(
    label_pos="center",
    is_label_show=True,
    label_text_color=None,
    is_legend_show=False,
    is_random=True
)
x = 10
y = 30
for i in range(len(total_age.columns)):
    pie.add(
      "", [str(total_age.columns[i])+"남성", str(total_age.columns[i])+"여성"],
        [total_age.iloc[0][i],total_age.iloc[1][i]],
        center = [x,y], radius = [25,40], **pie_style
    )
    if i < 4:
        x = x+25
    if i == 3:
        x = 10
        y = 75
    if i > 4:
        x = x+25
pie.render()
pie
```

### 1. 각 (연령별,지역별) 감염자 수와 그 비율을 나타낸 그래프는 서로 상이함을 파악했습니다.
- 각 연령별 확진자 수 그래프에서는 <a style="text-decoration:none; color:red;">**70,80**</a>대의 수치가 낮았지만, 연령별 인구 대비 그래프에서는 상대적으로 <a style="text-decoration:none; color:red;">**높은 비율**</a>을 차지했습니다.
- 지역의 경우 역시 확진자 수 그래프에서는 <a style="text-decoration:none; color:red;">**서울, 경기**</a> 외의 지역은 완만한 그래프를 그렸으나, 지역 인구 대비 그래프에서는 <a style="text-decoration:none; color:red;">**세종, 충남**</a>이 가장 큰 기울기를 기록했습니다.

### 2. 추가적으로 각 도시별 확진자/완치자 그래프를 그려보았습니다.
- 초기에 전염이 퍼진 도시의 경우 완치자의 비율이 **40-45%**를 차지했습니다.
- 반면 뒤늦게 감염된 도시는 아직 **30-35%**의 완치율을 보였습니다.

### 3. 각 연령별 성비에 대한 비교 그래프를 시각화 해보았습니다.
